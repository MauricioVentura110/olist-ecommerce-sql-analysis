```python
with open("olist_order_reviews_dataset.csv", encoding="utf-8") as f, \
     open("olist_order_reviews_dataset_clean.csv", "w", encoding="utf-8") as out: 
    header = next(f) # leer encabezado
    out.write(header) # escribir encabezado
    

    for i, line in enumerate(f, 2):
        # Primer filtro: asegurar que hay 6 comas, cada coma corresponde a una columna de esta tabla
        if line.count(",") == 6: 
            
            # Elemento 0 es el ID, se quitan los espacios al inicio y al final, así como los carecteres '"'
            split_line_0 = line.split(",")[0].strip(' "')
            # Elemento 5 es una columna de fecha y hora. split(" ") separa la fecha de la hora
            split_line_5 = line.split(",")[5].strip().split(" ")  
            # Elemento 6 es una columna de fecha y hora. split(" ") separa la fecha de la hora
            split_line_6 = line.split(",")[6].strip().split(" ") 
        
            # Segundo filtro: asegurar que el ID tenga 34 caracteres
            if len(split_line_0) != 32:
                continue
                
            # Tercer filtro: asegurar que el elemento 5 corresponde en efecto a un dato tipo fecha con hora
            if (split_line_5[0].count("-") != 2) or (split_line_5[1].count(":") !=2):
                continue
            
            # Cuarto filtro: asegurar que el elemento 6 corresponde en efecto a un dato tipo fecha con hora
            if (split_line_6[0].count("-") != 2) or (split_line_6[1].count(":") !=2):
                continue
                
            # Si se pasaron los filtros, la linea leída es útil para el análisis en la base de datos
            out.write(line)
```
